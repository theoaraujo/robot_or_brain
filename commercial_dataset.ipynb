# Creating commercial dataset

Performing check on image folders for the commercial dataset to (a) determine if dataset matches folder distribution and (b) create dataset containing the image unique ID and the predictions created by the finetuned clip model.

```python
import os
```

```python
import pandas as pd
```

```python
path = '/home/theo/robot_images/images_by_class/validation/'
```

```python
folders = os.listdir(path)
```

```python
df = []
for folder in folders:
    files = os.listdir(path+folder)
    for file in files:
        df.append({'folder': folder, 'filename': file})
    
```

```python
df = pd.DataFrame(df)
```

```python
df
```

## Merging with clip predictions

```python
predictions = pd.read_pickle('predictions_commercial.pkl')
```

```python
len(df), len(predictions)
```

```python
df = df.reset_index()
```

```python
df
```

```python
predictions
```

```python
results = df.merge(predictions)
```

```python
results
```

```python
results.to_csv('predictions_dataset.csv',index=False)
results.to_pickle('predictions_dataset.pkl')
```
